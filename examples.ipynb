# Civis matplotlib styles

This notebook shows how to use the matplotlib style functionality in `civis_mpl_style`.

## Using the matplotlib theme

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 100)
y = np.sqrt(x) + np.random.RandomState(42).rand(len(x))
with plt.style.context('civis'):
    fig, ax = plt.subplots()
    ax.plot(x, y, marker='o')
    plt.title("A plot of some data!")
    plt.show()
```

## Using single colors

```python
import civis_mpl_style
civis_mpl_style.add_to_mpl()

fig, ax = plt.subplots()
ax.hist(y, color='civis.darkyellow')
plt.show()
```

## Using the seaborn palette

```python
import matplotlib.pyplot as plt

import civis_mpl_style

with civis_mpl_style.seaborn_palette:
    fig, ax = plt.subplots()
    ax.hist(y, color='civis.blue')
    plt.show()
```
